```python
import matplotlib.pyplot as plt

# Lecture des données depuis le fichier
with open("bartbase.txt", "r") as file:
    lines = file.readlines()

epochs = []
loss = []

for line in lines:
    if "Epoch" in line:
        epoch = int(line.split(",")[0].split()[1])
        epochs.append(epoch)
        loss.append(float(line.split(",")[2].split(":")[1]))

# Création du graphique
plt.figure(figsize=(10, 6))
plt.plot(epochs, loss, marker='o', linestyle='-')

# Ajout de titres et de labels
plt.title('Loss curve by period')
plt.xlabel('Epochs')
plt.ylabel('Loss')

# Affichage du graphique
plt.grid(True)
plt.show()
```
